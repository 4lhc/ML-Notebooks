```python
# Importing the libraries 
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

## Load File

```python
# Importing the dataset
dataset = pd.read_csv('/content/pima-indians-diabetes.csv')
```

## Selecting Class Labels and feature set

```python
X = dataset.drop(columns='Diab')
print(X.head(), '\n')

y = dataset['Diab']
print(y.head())
```

## Precision, Recall, F1 score, ROC - Dummy Classifier for all features 

```python
from sklearn.dummy import DummyClassifier
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, roc_curve, auc, precision_recall_curve

X = dataset[['pregnant','plasma_glucose','age','bmi','skin_thickness', 'diab_pedigree', 'dia_BP']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
dummy = DummyClassifier(strategy="most_frequent", random_state =42)
dummy.fit(X_train, y_train)
dummy.score(X_test, y_test)

y_pred = dummy.predict(X_test)
print('Accuracy of dummy classifier on test set: {:.2f}'.format(dummy.score(X_test, y_test)))


confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)


y_pred_prob_lr = dummy.predict_proba(X_test)[:,1]
fpr_lr, tpr_lr, thresholds_lr = roc_curve(y_test, y_pred_prob_lr)
roc_auc_lr = auc(fpr_lr, tpr_lr)
precision_lr, recall_lr, th_lr = precision_recall_curve(y_test, y_pred_prob_lr)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_lr, tpr_lr, label='dummy classifier (area = %0.3f)' % roc_auc_lr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title(' Dummy Classifier - ROC curve for all features')
plt.legend(loc='best')
plt.show()
print('Precision:', precision_score(y_test, dummy.predict(X_test)))
print('Recall:', recall_score(y_test, dummy.predict(X_test)))
print('F1 score:', f1_score(y_test, dummy.predict(X_test)))
```

## SVM classification models with ‘rbf’ kernel - For all features

```python
from sklearn import svm
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score, roc_curve, auc, precision_recall_curve

X = dataset[['pregnant','plasma_glucose','age','bmi','skin_thickness', 'diab_pedigree', 'dia_BP']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
clf = svm.SVC(kernel='rbf', gamma='0.1', C='1') 
clf=svm.SVC(probability=True)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)

y_pred = clf.predict(X_test)
print('Accuracy of SVM model on test set: {:.2f}'.format(clf.score(X_test, y_test)))


confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)


y_pred_prob_lr = clf.predict_proba(X_test)[:,1]
fpr_lr, tpr_lr, thresholds_lr = roc_curve(y_test, y_pred_prob_lr)
roc_auc_lr = auc(fpr_lr, tpr_lr)
precision_lr, recall_lr, th_lr = precision_recall_curve(y_test, y_pred_prob_lr)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr_lr, tpr_lr, label='SVM (area = %0.3f)' % roc_auc_lr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title(' SVM - ROC curve for all features')
plt.legend(loc='best')
plt.show()
print('Precision:', precision_score(y_test, clf.predict(X_test)))
print('Recall:', recall_score(y_test, clf.predict(X_test)))
print('F1 score:', f1_score(y_test, clf.predict(X_test)))
```
